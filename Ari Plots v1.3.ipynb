```python
def temperatureFunction1(radius,Tr,metallicity):
    constantA = 5.6e-07
    constant1 = (1/constantA)*1.2e+04*(Tr**(-metallicity))
    result1 = constant1**((metallicity-6.5)*((14*(6.5-metallicity)**-2)+2*(1/(6.5-metallicity)))**(6.5-metallicity))*radius**(2/(6.5-metallicity))
    return result1

def temperatureFunction2(radius,Tr,Tm,metallicity):
    constantA = 5.6e-07
    constant2 = (1/constantA)*1.2e+04*(Tr**(-metallicity))*(Tm**(metallicity+1))
    result2 = (((169.0/252.0)*constant2)**(2.0/13.0))*radius**(4.0/13.0)
    return result2

def temperatureFunction3(radius,Tr,Tm,Tb,metallicity):
    constantA = 5.6e-07
    constant3 = (1/constantA)*1.2e+04*(Tr**(-metallicity))*(Tm**(metallicity+1))*(Tb**(-4.0/3.0))
    result3 = (((961.0/876.0)*constant3)**(6.0/31.0))*radius**(12.0/31.0)
    return result3
```
